# Matplotlib Basics

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Partitioning the real line

In order to plot a curve, we need a set of $x$ values and the corresponding $y$ values. Since x is the independent variable, we need to generate a list of values for x

```python
# np.linspace() produces 50 values by default

# 5 points between 0 and 1
np.linspace(0, 1, 5)
```

### Curve-1

Ploy $y = x$ for $x \in [0, 10]$. Sample 10 equally spaced

```python
x = np.linspace(0, 1, 10)
y = x

plt.scatter(x, y)
plt.plot(x, y)
```

**Note :** 
To supress these kind of outputs : `[<matplotlib.lines.Line2D at 0x1d833027340>]`, use semi-colon : 

`plt.plot(x, y);`

### Curve-2

 **Adding title to a plot**.

Plot $y = 5 - 3x$ for $x \in [-5, 5]$. Sample 20 equally spaced points in this intercal inclusive of the endpoints.

```python
np.linspace(-5, 5, 20)
y = 5 - 3*x

plt.scatter(x, y)
plt.title("$y = 5 - 3x$") # Title supports LaTeX
plt.plot(x, y);
```

## Curve-3

**Adding x-axis and y-axis with gridlines for a plot**

Plot the graph $y = x^2$. Try out four different number of sample : 
- Sample 5 equally spaced points in this intercal inclusive of the endpoints
- Sample 10 equally spaced points in this intercal inclusive of the endpoints
- Sample 20 equally spaced points in this intercal inclusive of the endpoints
- Sample 50 equally spaced points in this intercal inclusive of the endpoints



**Note** : `plt.plot(x, y)` just draws straight-lines between the consecutive points

```python
x = np.linspace(-1, 1, 50)
y = x**2

plt.scatter(x, y)
plt.plot(x, y)
plt.title("$y = x^2$")

plt.axhline(color="black");
plt.axvline(color="black");

plt.grid(True)
```

### Curve-4

**Coloring the lines on the graph and adding a legend**

Plot $y = 3$ and $x = 5$ on the same plot. Color the first one green and the second one red. Limit your plot to $x \in [0,6]$ and $y \in [0,6]$. 

- Label the x and y axes
- Add a label
- Add a legend

```python
# Plot y = 3

x = np.linspace(0, 6)
n = x.shape[0]
y = np.ones(n) * 3

plt.plot(x, y, color="green", label="$y = 3$")

# Plot x = 5
x = np.ones(n) * 5
y = np.linspace(0, 6)

plt.plot(x, y, color="red", label="$x = 5$")

plt.axhline(color="black")
plt.axvline(color="black")

plt.grid(True, color="#d4deff")

plt.title("$x = 5, y = 3$")
plt.legend();
```

## Subplots

We use `plt.subplot(m, n, index)` to get `m x n` subplots. To access each of them, we use the `index` parameter. The index goes from left to right top to bottom.

Plot the following curves, each in a separate plot : 

- $y = \ln x, x \in [e^2, e^3]$
- $y = e^x, x \in [0, 3]$
- $y = \sin(x), x \in [0, 2\pi]$
- $y = \frac {1} {x} , x \in [0, 5]$

**Note**

- `plt.rcParams` : You can dynamically change the default rc (runtime configuration) settings in a python script or interactively from the python shell. All rc settings are stored in a dictionary-like variable called matplotlib.rcParams, which is global to the matplotlib package

```python
plt.rcParams["figure.figsize"] = [8, 8]
plt.rcParams["lines.linewidth"] = 2

# We can specify a list of colors to cycle automatically too
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=["#1f77b4"]) 


# Plot y = ln(x)
plt.subplot(2, 2, 1)

x = np.linspace(1 / np.e**2, np.e**3)
y = np.log(x)

plt.plot(x, y)
plt.title("$y = \ln x$")

# Plot y = e^x
plt.subplot(2, 2, 2)

x = np.linspace(0, 3)
y = np.exp(x)

plt.plot(x, y)
plt.title("$y = e^x$")

# Plot y = e^x
plt.subplot(2, 2, 2)

x = np.linspace(0, 3)
y = np.exp(x)

plt.plot(x, y)
plt.title("$y = e^x$")

# Plot y = sin(x)
plt.subplot(2, 2, 3)

x = np.linspace(0, 2 * np.pi)
y = np.sin(x)

plt.plot(x, y)
plt.title("$y = e^x$")

# Plot y = 1/x
plt.subplot(2, 2, 4)

x = np.linspace(0.1, 5) # 0-divsion error. So, we start from 0.001
y = 1 / x

plt.plot(x, y)
plt.title("$y = 1/x$");
```
